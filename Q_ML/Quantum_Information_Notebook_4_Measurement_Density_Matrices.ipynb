## Quantum Measurement, Density Matrices, and Mixed States


# Quantum Measurement

For a state

$$
|\psi\rangle=\alpha|0\rangle+\beta|1\rangle,
$$

measurement in the computational basis gives

$$
P(0)=|\alpha|^2,\qquad
P(1)=|\beta|^2,
$$

with

$$
|\alpha|^2+|\beta|^2=1.
$$

After measurement, the state collapses to the observed basis state.



# Projective Measurement

Projectors satisfy

$$
P_i^2=P_i,\qquad P_i^\dagger=P_i.
$$

For the computational basis,

$$
P_0=|0\rangle\langle0|,
\qquad
P_1=|1\rangle\langle1|.
$$

Probability of outcome \(i\):

$$
p_i=\langle\psi|P_i|\psi\rangle.
$$



# Density Matrix

Instead of describing a state using a ket, we can use

$$
\rho=|\psi\rangle\langle\psi|.
$$

For

$$
|\psi\rangle=\frac{|0\rangle+|1\rangle}{\sqrt2},
$$

$$
\rho=
\frac12
\begin{bmatrix}
1&1\\
1&1
\end{bmatrix}.
$$

Properties:

- Hermitian
- Positive semidefinite
- Trace = 1



# Pure vs Mixed States

A **pure state** contains complete quantum information.

Example:

$$
|0\rangle.
$$

A **mixed state** represents a statistical mixture.

Example:

50% chance of $|0\rangle$

50% chance of $|1\rangle$

Density matrix

$$
\rho=
\frac12
\begin{bmatrix}
1&0\\
0&1
\end{bmatrix}.
$$

A useful test:

$$
\mathrm{Tr}(\rho^2)=
\begin{cases}
1 & \text{Pure}\\
<1 & \text{Mixed}
\end{cases}
$$


# Partial Trace

The partial trace removes one subsystem while preserving information about the remaining subsystem.

Example:

```
AB system
 │
 ├── Trace over B
 │
 ▼
State of A
```

This is essential when studying entanglement.



# Bloch Vector Representation

Any single-qubit density matrix can be written as

$$
\rho=
\frac12
\left(
I+
r_xX+
r_yY+
r_zZ
\right),
$$

where

$$
\vec r=(r_x,r_y,r_z)
$$

is the Bloch vector.

- Pure states lie on the surface.
- Mixed states lie inside the sphere.


````python
from qiskit.quantum_info import Statevector, DensityMatrix

psi = Statevector.from_label("0")
rho = DensityMatrix(psi)

print("Statevector:")
print(psi)

print("\nDensity Matrix:")
print(rho.data)
````

````python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

qc = QuantumCircuit(1,1)
qc.h(0)
qc.measure(0,0)

print(qc.draw())

sim = AerSimulator()
result = sim.run(qc, shots=1000).result()
print(result.get_counts())
````
