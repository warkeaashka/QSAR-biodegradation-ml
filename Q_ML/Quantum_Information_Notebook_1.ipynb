# Dirac (Bra-Ket) Notation

A general qubit is

$
|\psi\rangle=\alpha|0\rangle+\beta|1\rangle
$

where

$
|\alpha|^2+|\beta|^2=1.
$

The measurement probabilities are

- $P(0)=|\alpha|^2\$


- $P(1)=|\beta|^2\$



# Worked Example

Given

$
|\psi\rangle=\frac{1}{\sqrt2}|0\rangle+\frac{1}{\sqrt2}|1\rangle
$


then

$
P(0)=\frac12,\qquad
P(1)=\frac12.
$

The state is perfectly balanced.


````python

from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

qc = QuantumCircuit(1,1)
qc.h(0)
qc.measure(0,0)

print(qc.draw())

sim = AerSimulator()
result = sim.run(qc, shots=1000).result()
print(result.get_counts())
````


# Understanding the Circuit

# Measurement

Measurement collapses the quantum state.

Before measurement:

$
|\psi\rangle=\alpha|0\rangle+\beta|1\rangle
$

After measurement:

- either $|0\rangle$
- or $|1\rangle$
