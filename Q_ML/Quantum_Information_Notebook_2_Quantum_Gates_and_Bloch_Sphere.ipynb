## Quantum Gates and the Bloch Sphere


## Quantum Gates

Quantum gates transform quantum states.

$
|\psi'\rangle = U|\psi\rangle
$

where \(U\) is a **unitary matrix** satisfying

$
U^\dagger U = I.
$

This preserves total probability.



## Important Single-Qubit Gates

### Pauli-X (Quantum NOT)

$
X=\begin{bmatrix}0&1\\1&0\end{bmatrix}
$

Swaps $|0\rangle\leftrightarrow|1\rangle$

### Pauli-Y

$
Y=\begin{bmatrix}
0&-i\\
i&0
\end{bmatrix}
$

### Pauli-Z

$
Z=\begin{bmatrix}
1&0\\
0&-1
\end{bmatrix}
$

Flips the phase of $|1\rangle$



## Hadamard Gate

$
H=\frac{1}{\sqrt2}
\begin{bmatrix}
1&1\\
1&-1
\end{bmatrix}
$

Creates equal superposition:

$
H|0\rangle=
\frac{|0\rangle+|1\rangle}{\sqrt2}.
$



## Bloch Sphere

Every pure qubit can be represented as

$
|\psi\rangle=
\cos\frac{\theta}{2}|0\rangle+
e^{i\phi}\sin\frac{\theta}{2}|1\rangle.
$

ASCII sketch

```
        |0>
         ●
        /|
       / |
------●--+-------
      \  |
       \ |
        \|
         ●
        |1>
```

Rotations move the state around the sphere.


````python

from qiskit import QuantumCircuit

from qiskit_aer import AerSimulator

qc = QuantumCircuit(1,1)

qc.h(0)
qc.z(0)
qc.measure(0,0)

print(qc.draw())

sim = AerSimulator()
counts = sim.run(qc, shots=1000).result().get_counts()
print(counts)
````

````python

from qiskit import QuantumCircuit

qc = QuantumCircuit(1)

qc.rx(3.14159/2,0)
qc.ry(3.14159/4,0)
qc.rz(3.14159/3,0)

print(qc.draw())
````


## Worked Example

Starting state $|0\rangle$

Apply:

1. Hadamard
2. Pauli-X

Step 1:

$
H|0\rangle=
\frac{|0\rangle+|1\rangle}{\sqrt2}
$

Step 2:

$
XH|0\rangle=
\frac{|1\rangle+|0\rangle}{\sqrt2}
=
\frac{|0\rangle+|1\rangle}{\sqrt2}.
$

The probabilities remain unchanged.
