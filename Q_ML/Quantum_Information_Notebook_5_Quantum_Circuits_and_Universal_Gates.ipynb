## Quantum Circuits and Universal Gate Sets


# The Quantum Circuit Model

A quantum algorithm is represented as a sequence of quantum gates applied to qubits.

```
|q0⟩ ──H────■────────M──
            │
|q1⟩ ───────X────────M──
```

Each horizontal line is a qubit.
Time flows from left to right.



# Reversible Computation

Unlike most classical logic gates, quantum gates are reversible.

If

$$
|\psi'\rangle = U|\psi\rangle,
$$

then

$$
|\psi\rangle = U^\dagger |\psi'\rangle,
$$

where $U^\dagger$ is the inverse (conjugate transpose) of $U\$.



# Universal Gate Sets

A small collection of gates can approximate any quantum computation.

Common universal sets include:

- {H, T, CNOT}
- {RX, RY, RZ, CNOT}
- {U3, CNOT} (legacy naming in Qiskit)

The celebrated Solovay–Kitaev theorem states that arbitrary unitary operations can
be approximated efficiently using a finite universal gate set.



# Multi-Control Gates

## Toffoli (CCX)

```
q0 ──●──────
      │
q1 ──●──────
      │
q2 ──⊕──────
```

The target flips only when both controls are 1.

## Fredkin (CSWAP)

```
q0 ──●────────
      │
q1 ──×────────
      │
q2 ──×────────
```

The target qubits are swapped if the control is 1.



# Circuit Decomposition

Complex gates are often decomposed into simpler gates supported by hardware.

Example:

$$
U \approx H \rightarrow T \rightarrow H \rightarrow CNOT.
$$

Decomposition is essential because real quantum devices implement only a limited
native gate set.


````python

from qiskit import QuantumCircuit

qc = QuantumCircuit(3)
qc.h(0)
qc.cx(0,1)
qc.ccx(0,1,2)

print(qc.draw())
````

````python

from qiskit import transpile
from qiskit_aer import AerSimulator
from qiskit import QuantumCircuit

qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0,1)

backend = AerSimulator()
optimized = transpile(qc, backend=backend, optimization_level=3)

print("Original circuit:")
print(qc.draw())

print("\nOptimized circuit:")
print(optimized.draw())
````


# Worked Example

Goal: Create an entangled pair.

1. Initialize $|00\rangle$
2. Apply Hadamard to the first qubit.
3. Apply CNOT.

Result:

$$
|\Phi^+\rangle=
\frac{|00\rangle+|11\rangle}{\sqrt2}.
$$

This simple two-gate circuit is a building block for many quantum algorithms.


- Keep circuits as shallow as possible.
- Minimize two-qubit gates because they are typically noisier.
- Reuse qubits efficiently.
- Transpile circuits for the target hardware.
