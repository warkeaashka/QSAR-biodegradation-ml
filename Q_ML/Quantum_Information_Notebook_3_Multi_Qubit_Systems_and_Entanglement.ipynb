## Multi-Qubit Systems, Tensor Products, and Entanglement



# From One Qubit to Two Qubits

A single qubit has two basis states:

$$
|0\rangle,\quad |1\rangle
$$

Two qubits have four basis states:

$$
|00\rangle,\ |01\rangle,\ |10\rangle,\ |11\rangle.
$$

In general, \(n\) qubits require \(2^n\) basis states.



# Tensor Product

Two independent qubits combine using the tensor product.

Example:

$$
|0\rangle\otimes|1\rangle = |01\rangle.
$$

If

$$
|\psi\rangle=\alpha|0\rangle+\beta|1\rangle,\qquad
|\phi\rangle=\gamma|0\rangle+\delta|1\rangle,
$$

then

$$
|\psi\rangle\otimes|\phi\rangle
=
\alpha\gamma|00\rangle+
\alpha\delta|01\rangle+
\beta\gamma|10\rangle+
\beta\delta|11\rangle.
$$

# Bell States

The four Bell states are maximally entangled.

$$
|\Phi^+\rangle=
\frac{|00\rangle+|11\rangle}{\sqrt2}
$$

$$
|\Phi^-\rangle=
\frac{|00\rangle-|11\rangle}{\sqrt2}
$$

$$
|\Psi^+\rangle=
\frac{|01\rangle+|10\rangle}{\sqrt2}
$$

$$
|\Psi^-\rangle=
\frac{|01\rangle-|10\rangle}{\sqrt2}
$$


Entangled particles cannot always be described independently.

Example:

```
Measure first qubit
        вҶ“
Second qubit becomes correlated immediately.
```

This correlation does **not** allow faster-than-light communication, but it is a key resource in quantum information.


````python

from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

qc = QuantumCircuit(2,2)

qc.h(0)
qc.cx(0,1)
qc.measure([0,1],[0,1])

print(qc.draw())

sim = AerSimulator()
counts = sim.run(qc, shots=1024).result().get_counts()

print(counts)
````
