# What is a Quantum Gate?

A **quantum gate** is the quantum equivalent of a **classical logic gate**. Just as classical logic gates perform operations on bits, quantum gates perform operations on **qubits**, changing their quantum states according to the principles of quantum mechanics.

### Classical vs Quantum

- **Classical computers** manipulate **bits**, which can only exist in one of two states: **0** or **1**.
- **Quantum computers** manipulate **qubits**, which can exist in a **superposition** of both states simultaneously:

$
|\psi\rangle = \alpha|0\rangle + \beta|1\rangle
$

where:

- $\alpha$ and $\beta$ are complex probability amplitudes,
- $|\alpha|^2 + |\beta|^2 = 1$

Quantum gates modify these probability amplitudes, thereby changing the state of a qubit.

### Reversibility of Quantum Gates

Unlike classical logic gates (such as AND or OR), **every quantum gate is reversible**. This means that the original quantum state can always be recovered by applying the inverse operation.

Mathematically, every quantum gate is represented by a **unitary matrix** \(U\), which satisfies the following condition:

$
U^{\dagger}U = I
$

where:

- $U^{\dagger}$ = Conjugate transpose (Hermitian adjoint) of \(U\)
- $I\$ = Identity matrix

This property guarantees that:

- Quantum information is never lost.
- The total probability remains equal to 1.
- Every quantum operation has an inverse.

----

- Quantum gates operate on **qubits**, not classical bits.
- They are represented by **unitary matrices**.
- Quantum gates are **reversible**, unlike many classical logic gates.
- They preserve quantum information and probability throughout computation.

```python
import numpy as np

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

from qiskit.visualization import plot_bloch_multivector
from qiskit.visualization import plot_histogram

from qiskit_aer import AerSimulator
```

```python
qc = QuantumCircuit(1)

qc.draw("mpl")
```

The qubit is initially in the **ground state** $|0\rangle$, represented in vector form as

$$
|0\rangle =
\begin{bmatrix}
1 \\
0
\end{bmatrix}
$$


This means that the probability of measuring the qubit in the state $|0\rangle$ is **1**, while the probability of measuring it in the state $|1\rangle$ is **0**.

In quantum computing, $|0\rangle$ is the default initial state of a qubit before any quantum gate is applied.

## Identity Gate (I Gate)
#### Purpose

The identity gate does absolutely nothing.

Think of it as multiplying a number by 1.

State remains unchanged.

Matrix: 
$$
\begin{bmatrix}
1 & 0\\
0 & 1
\end{bmatrix}

```python
qc = QuantumCircuit(1)

qc.id(0)

state = Statevector.from_instruction(qc)

print(state)
qc.draw("mpl")
```

# Pauli-X Gate (X Gate)

The **Pauli-X Gate**, also called the **X Gate** or **Quantum NOT Gate**, is one of the most fundamental single-qubit quantum gates.

It behaves similarly to the **classical NOT gate**, as it flips the state of a qubit.

The Pauli-X gate swaps the computational basis states:

$
|0\rangle \rightarrow |1\rangle
$

$
|1\rangle \rightarrow |0\rangle
$ 


The Pauli-X gate is represented by the following unitary matrix:

$
X =
\begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix}
$


Initial state:

$
|0\rangle =
\begin{bmatrix}
1 \\
0
\end{bmatrix}
$

Apply the X gate:

$
X|0\rangle =
\begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix}
\begin{bmatrix}
1 \\
0
\end{bmatrix}
=
\begin{bmatrix}
0 \\
1
\end{bmatrix}
=
|1\rangle
$


Initial state:

$
|1\rangle =
\begin{bmatrix}
0 \\
1
\end{bmatrix}
$

Apply the X gate:

$
X|1\rangle =
\begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix}
\begin{bmatrix}
0 \\
1
\end{bmatrix}
=
\begin{bmatrix}
1 \\
0
\end{bmatrix}
=
|0\rangle
$

- Known as the **Quantum NOT Gate**.
- Flips the state of a qubit:
  - $|0\rangle \leftrightarrow |1\rangle$
- Represented by a **2 × 2 unitary matrix**.
- It is **reversible**, meaning:

$
X^2 = I
$

Applying the Pauli-X gate twice returns the qubit to its original state.

```python
qc = QuantumCircuit(1)

qc.x(0)

state = Statevector.from_instruction(qc)

print(state)
qc.draw("mpl")
```

# Pauli-Y Gate (Y Gate)

The **Pauli-Y Gate**, also called the **Y Gate**, changes both the **bit** and the **phase** of the qubit. It introduces a phase factor of **\(i\)** or **\(-i\)** while flipping the state.

The Pauli-Y gate is represented by the following unitary matrix:

$
Y =
\begin{bmatrix}
0 & -i \\
i & 0
\end{bmatrix}
$

where

$
i=\sqrt{-1}
$

is the imaginary unit.

#### State Transformations

Applying the Y gate results in:

$
|0\rangle \rightarrow i|1\rangle
$

$
|1\rangle \rightarrow -i|0\rangle
$

Notice that the qubit **changes state** and also acquires a **complex phase factor**.

Applying Y to $|0\rangle$

Initial state:

$
|0\rangle =
\begin{bmatrix}
1 \\
0
\end{bmatrix}
$

Apply the Y gate:

$
Y|0\rangle =
\begin{bmatrix}
0 & -i \\
i & 0
\end{bmatrix}
\begin{bmatrix}
1 \\
0
\end{bmatrix}
=
\begin{bmatrix}
0 \\
i
\end{bmatrix}
=
i|1\rangle
$


Applying Y to $|1\rangle$

Initial state:

$
|1\rangle =
\begin{bmatrix}
0 \\
1
\end{bmatrix}
$

Apply the Y gate:

$
Y|1\rangle =
\begin{bmatrix}
0 & -i \\
i & 0
\end{bmatrix}
\begin{bmatrix}
0 \\
1
\end{bmatrix}
=
\begin{bmatrix}
-i \\
0
\end{bmatrix}
=
-i|0\rangle
$

------------------------------

- Flips the qubit state **and** changes its phase.
- Introduces a phase factor of **\(i\)** or **\(-i\)**.
- Represented by a **2 × 2 unitary matrix**.
- Like all quantum gates, it is **reversible**.
- Applying the Y gate twice returns the original state:

$
Y^2 = I
$

where $I\$ is the identity matrix.

```python
qc = QuantumCircuit(1)

qc.y(0)
state = Statevector.from_instruction(qc)

print(state)
qc.draw("mpl")
```

# Pauli-Z Gate (Z Gate)

The **Pauli-Z Gate**, also called the **Z Gate** or **Phase Flip Gate**, changes only the **phase** of the $|1\rangle$ state by multiplying it by **−1**.

The Pauli-Z gate is represented by the following unitary matrix:

$
Z =
\begin{bmatrix}
1 & 0 \\
0 & -1
\end{bmatrix}
$

### State Transformations

Applying the Z gate gives:

$
|0\rangle \rightarrow |0\rangle
$

$
|1\rangle \rightarrow -|1\rangle
$

Notice that:

- The state $|0\rangle$ remains unchanged.
- The state $|1\rangle$ keeps its value but acquires a phase factor of **−1**.

Applying Z to $|0\rangle$

Initial state:

$
|0\rangle =
\begin{bmatrix}
1 \\
0
\end{bmatrix}
$

Apply the Z gate:

$
Z|0\rangle =
\begin{bmatrix}
1 & 0 \\
0 & -1
\end{bmatrix}
\begin{bmatrix}
1 \\
0
\end{bmatrix}
=
\begin{bmatrix}
1 \\
0
\end{bmatrix}
=
|0\rangle
$

The qubit remains unchanged.

Applying Z to $|1\rangle$

Initial state:

$
|1\rangle =
\begin{bmatrix}
0 \\
1
\end{bmatrix}
$

Apply the Z gate:

$
Z|1\rangle =
\begin{bmatrix}
1 & 0 \\
0 & -1
\end{bmatrix}
\begin{bmatrix}
0 \\
1
\end{bmatrix}
=
\begin{bmatrix}
0 \\
-1
\end{bmatrix}
=
-|1\rangle
$

The qubit remains in the $|1\rangle$ state but gains a phase factor of **−1**.

_______________________________

- Also known as the **Phase Flip Gate**.
- Does **not** flip the qubit state.
- Changes only the **phase** of the \(|1\rangle\) state.
- Represented by a **2 × 2 unitary matrix**.
- It is **reversible**, and

$
Z^2 = I
$

where $I\$ is the identity matrix.

```python
qc = QuantumCircuit(1)

qc.z(0)

qc.draw("mpl")
```

## Hadamard Gate (H Gate)

The **Hadamard gate** is one of the most important gates in quantum computing.
It creates **superposition**.

The matrix representation of Hadamard Gate is given by:

$
H=\frac{1}{\sqrt{2}}
\begin{bmatrix}
1 & 1\\
1 & -1
\end{bmatrix}
$

Apply H to $|0\rangle$

Result

$
H|0\rangle=
\frac{|0\rangle+|1\rangle}{\sqrt{2}}
$


- 50% probability of measuring **0**
- 50% probability of measuring **1**

```python
qc = QuantumCircuit(1)

qc.h(0)

qc.draw("mpl")
```
